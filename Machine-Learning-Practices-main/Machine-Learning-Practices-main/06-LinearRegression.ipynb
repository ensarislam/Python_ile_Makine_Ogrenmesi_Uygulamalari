```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# veri olu≈üturma
X = np.random.rand(100,1)
y = 3 + 4 * X + np.random.rand(100,1)    # y=3+4x
```

```python
plt.scatter(X,y)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
plt.figure()
plt.scatter(X,y)
plt.plot(X, lin_reg.predict(X), color = "red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lineer Regresyon Modeli")
```

```python
# y=3+4x  --> y = a0 + a1x
a1 = lin_reg.coef_[0][0]
print("a1: ",a1)

a0 = lin_reg.intercept_[0]
print("a0: ",a0)

for i in range(100):
    y_ = a0 + a1 * X
    plt.plot(X , y_ , color = "green" , alpha = 0.7)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
```

```python
diabetes = load_diabetes()
```

```python
diabetes.feature_names
```

```python
diabetes_X , diabetes_y = load_diabetes(return_X_y=True)   # X:feature  y:target
```

```python
diabetes_X = diabetes_X[: , np.newaxis , 2]
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
```

```python
diabetes_y_train = diabetes_y[:-20]   
diabetes_y_test = diabetes_y[-20:]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(diabetes_X_train,diabetes_y_train)
```

```python
diabetes_y_pred = lin_reg.predict(diabetes_X_test)
```

```python
mse = mean_squared_error(diabetes_y_test,diabetes_y_pred)
mse
```

```python
r2 = r2_score(diabetes_y_test,diabetes_y_pred)
r2
```

```python
plt.scatter(diabetes_X_test, diabetes_y_test, color = "black")
plt.plot(diabetes_X_test, diabetes_y_pred, color = "blue")
```
