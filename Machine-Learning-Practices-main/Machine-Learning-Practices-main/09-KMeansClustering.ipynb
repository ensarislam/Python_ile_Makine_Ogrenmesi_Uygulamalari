```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# veri seti olu≈üturma
X , _ = make_blobs(n_samples=300 , centers = 4 , cluster_std=0.6 , random_state=42)
```

```python
plt.figure()
plt.scatter(X[:,0], X[:,1])
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
# KMeans modelini X veri seti ile eƒüittikten sonra veri setine ait 0,1,2,3 olucak ≈üekilde 4 tane k√ºme olu≈üturur. 
# A≈üaƒüƒ±da veri setindeki her bir verinin hangi k√ºmeye ait olduƒüunu g√∂sterir.
```

```python
labels = kmeans.labels_
labels
```

```python
plt.figure()
plt.scatter(X[:,0], X[:,1] , c = labels , cmap = "viridis")

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1] , c = "red" , marker = "X")
```
