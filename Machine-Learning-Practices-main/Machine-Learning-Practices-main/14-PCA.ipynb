```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# iris veri seti ile 3 farklı türe ait sınıflandırma problemini çözeriz.
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
pca = PCA(n_components=2)
```

```python
# iris veri setini kullanarak PCA analizi ile boyut indirgeme yaparız ve principal componentları(PC1,PC2) buluruz.
```

```python
X_pca = pca.fit_transform(X)
df = pd.DataFrame(X_pca)
df
```

```python
# PC'ları kullanarak sınıflandırma problemini çözeriz.
```

```python
plt.figure()

for i in range(len(iris.target_names)):
    plt.scatter(X_pca[y==i,0],X_pca[y==i,1] , label = iris.target_names[i])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Iris Dataset")
plt.legend()
```

```python
# iris veri setini kullanarak PCA analizi ile 3 adet principal component(temel bileşen) tanımlarız. Bu PC'ları kullanarak
# sınıflandırma problemini çözeriz.
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
```

```python
df = pd.DataFrame(X_pca)
df
```

```python
fig = plt.figure(1, figsize = (8,6))
ax = fig.add_subplot(111 , projection = "3d" , elev=-150, azim = 110)

ax.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2], c = y ,s=40)
ax.set_title("First Three PCA Dimesions of Iris Dataset")
ax.set_xlabel("1st Eigenvector")
ax.set_ylabel("2st Eigenvector")
ax.set_zlabel("3st Eigenvector")
```
