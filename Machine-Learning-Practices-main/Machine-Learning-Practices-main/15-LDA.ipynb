```python
from sklearn.datasets import fetch_openml
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
mnist = fetch_openml("mnist_784",version=1)
```

```python
X = mnist.data
y = mnist.target.astype(int)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X,y)
```

```python
# mnist veri setini kullanarak LDA analizi ile boyut indirgeme yaptık ve görselleştirdik.
```

```python
plt.figure()
plt.scatter(X_lda[:,0],X_lda[:,1], c=y , cmap="tab10" , alpha = 0.6)
plt.title("LDA of MNIST Dataset")
plt.xlabel("LD1")
plt.ylabel("LD2")
plt.colorbar(label = "Digits")
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
```

```python
# iris veri setini kullanarak PCA ve LDA analizi ile boyut indirgeme yaparak 2 adet principal component(temel bileşen) 
# tanımlarız. Bu PC'ları kullanarak sınıflandırma problemini çözeriz.
```

```python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X,y)
```

```python
colors = ["red", "blue" , "green"]

plt.figure()
for color, i ,target_name in zip(colors, [0,1,2], target_names):
    plt.scatter(X_pca[y==i,0],X_pca[y==i,1], color = color , alpha=0.8, label = target_name)
plt.title("PCA of Iris Dataset")
plt.legend()

plt.figure()
for color , i , target_name in zip(colors , [0,1,2], target_names):
    plt.scatter(X_lda[y == i,0],X_lda[y == i,1], color=color , alpha=0.8, label = target_name)
plt.title("LDA of Iris Dataset")
plt.legend()
```
